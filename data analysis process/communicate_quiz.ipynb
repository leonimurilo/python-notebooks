```python
# imports and load data
import pandas as pd

df = pd.read_csv('store_data.csv')
```

```python
# explore data
df.hist(figsize=(8,8));
```

```python
df.tail()
```

```python
# sales for the last month
df.iloc[196:, 1:].sum().plot(kind='bar');
```

```python
# average sales
df.mean().plot(kind='bar')
```

```python
# sales for the week of March 13th, 2016
df['week'] = pd.to_datetime(df['week'])
df[df['week'] == '2016-03-13'].iloc[:,1:].plot(kind='bar')
```

```python
# sales for the lastest 3-month periods
# each store:
df[df['week'] >= '2017-12-01'].iloc[:,1:].plot()
```

```python
# all stores
df[df['week'] >= '2017-12-01'].iloc[:,1:].sum().plot(kind='pie', autopct='%1.0f%%');
```
